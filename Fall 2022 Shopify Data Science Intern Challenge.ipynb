## Import libraries and loading data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('2019 Winter Data Science Intern Challenge Data Set.csv')
```

## Box plot

```python
# Boxplot of Amount
bp = plt.boxplot(df['order_amount'], labels=['Order Amount'])
plt.title('Boxplot of Order Amount')
plt.ylabel('Dollar Value of Order')
plt.show()
```

## Histograms

```python
# Histogram of Order Amount
hist = plt.hist(df['order_amount'],bins=100)
plt.title('Histogram of Order Amount')
plt.xlabel('Order Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram of Order Amount with Order Amounts below AOV
hist = plt.hist(df[df.order_amount < 3145.13]['order_amount'],bins=100)
plt.title('Histogram of Order Amount below AOV')
plt.xlabel('Order Amount')
plt.ylabel('Frequency')
plt.show()

# Number of Orders with Order_Amount less than AOV
num_of_orders_under_AOV = len(df[df.order_amount < 3145.13])
print(num_of_orders_under_AOV)
```

## Median Order Value (MOV)

```python
# Median Order Value (MOV)
median = df['order_amount'].median()
print(median)
```
